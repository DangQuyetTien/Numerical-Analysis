```python
def multiply_by_x_minus_c(a, c):
    """
    Thực hiện nhân đa thức có hệ số trong mảng 'a' với (x - c).
    
    Args:
        a (list): Mảng chứa các hệ số của đa thức ban đầu, 
                  sắp xếp từ bậc cao nhất đến thấp nhất.
        c (float): Giá trị c.

    Returns:
        list: Mảng 'b' chứa các hệ số của đa thức kết quả.
    """
    # n là bậc của đa thức ban đầu
    n = len(a) - 1
    
    # B1: Khởi tạo mảng b chứa n+2 số 0, vì đa thức mới sẽ có bậc n+1
    b = [0.0] * (n + 2)
    
    # Gán hệ số cho bậc cao nhất (bậc n+1)
    # Trong Python, đây là phần tử đầu tiên b[0]
    b[0] = float(a[0])
    
    # Gán hệ số cho bậc thấp nhất (bậc 0)
    # Trong Python, đây là phần tử cuối cùng b[n+1]
    b[n+1] = -a[n] * c
    
    # B2: Dùng vòng lặp để tính các hệ số ở giữa
    # Vòng lặp for j từ 1 đến n (bao gồm n)
    # b[j] tương ứng với hệ số của x^(n+1-j)
    for j in range(1, n + 1):
        # Công thức b_i = a_{i-1} - c*a_i khi chuyển sang chỉ số mảng 0..n
        # sẽ tương đương với b[j] = a[j] - c*a[j-1]
        b[j] = a[j] - c * a[j-1]
        
    # B3: Trả về mảng kết quả b
    return b

# --- Ví dụ sử dụng ---
# Cho đa thức P(x) = 3x^2 + 2x - 5
# Nhân với (x - 4)
# Kết quả mong đợi: (3x^2 + 2x - 5)(x-4) = 3x^3 -12x^2 + 2x^2 -8x -5x + 20
#                  = 3x^3 - 10x^2 - 13x + 20
# Hệ số kết quả phải là [3, -10, -13, 20]

a = [3, 2, -5]  # Hệ số của P(x)
c = 4

b = multiply_by_x_minus_c(a, c)

print(f"Mảng hệ số ban đầu a = {a}")ơ]
print(f"Nhân với (x - {c})")
print(f"Mảng hệ số kết quả b = {b}") 
```
