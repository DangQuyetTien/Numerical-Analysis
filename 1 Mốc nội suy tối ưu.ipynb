Tìm n mốc nội suy tối ưu trong đoạn [ a; b ]

```python
import math


# Khởi tạo các giá trị ban đầu
a = 2
b = 5
n = 9

# Tạo một danh sách (list) rỗng để lưu trữ các mốc nội suy
x = [] 

# Tính toán các hằng số c và d để đơn giản hóa công thức
c = a - b
d = a + b

# Vòng lặp để tính toán các mốc nội suy Chebyshev
# Vòng lặp chạy từ i = 0 đến n
for i in range(n):
    # Tính giá trị của từng mốc nội suy theo công thức và thêm vào danh sách x
    node = 1/2 * (c * math.cos((2 * i + 1) * math.pi / (2 * (n))) + d)
    x.append(node)

# Định dạng lại danh sách x để mỗi số có 5 chữ số sau dấu phẩy
x_formatted = [f"{val:.5f}" for val in x]

# In kết quả ra màn hình
print(f"{n} mốc nội suy trong đoạn [{a},{b}] là:\n{x_formatted}")
```
