```python
import pandas as pd

def read_excel_column_to_list_string(file_path, sheet_name='Sheet1', col_to_read=1):
    """Đọc một cột từ file Excel và trả về dưới dạng chuỗi định dạng list."""
    try:
        df = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
        # Lấy dữ liệu từ cột được chỉ định (trừ 1 vì index cột cũng từ 0)
        col_data = df.iloc[:, col_to_read - 1].dropna().astype(int)
        data_list = col_data.tolist()
        return str(data_list)
    except FileNotFoundError:
        return f"Lỗi: Không tìm thấy file tại '{file_path}'"
    except Exception as e:
        return f"Đã xảy ra lỗi: {e}"

# --- Cách sử dụng cho cột ---
# Giả sử dữ liệu nằm ở cột A (cột thứ 1)
file_name = "input.xlsx"
result_string_col = read_excel_column_to_list_string(file_name, col_to_read=1)
print("\nKết quả đọc từ cột A:")
print(result_string_col)
```
