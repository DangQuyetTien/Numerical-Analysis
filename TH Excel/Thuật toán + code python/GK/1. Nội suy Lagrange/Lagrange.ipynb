```python
import numpy as np
import pandas as pd
```

```python
def doc_input(ten_file):
    # tra ve gia tri cua x va y tu file input.xlsx
    # doc file input.xlsx su dung pandas
    data = pd.read_excel(ten_file)
    
    # doc du lieu cua x va y
    x = data.iloc[:, 0]  # Assuming x is in the first column
    y = data.iloc[:, 1]  # Assuming y is in the second column
    print(type(y))
    
    # xu ly du lieu cua x va y
    x = np.array(x, dtype=float)
    #x = x.values
    if y.isnull().all():  # Check if y is empty or contains NaNs
        #y = f(x)  # Assuming f(x) is a function defined elsewhere
        y = np.array(y, dtype=float)
        #y = y.values

    else:
        y = np.array(y, dtype=float)
        #y = y.values

    return x, y
    
def hoocnerNhan(x):
    a = [1, -x[0]]
    for i in range(1, len(x)):
        a.append(0)
        temp = a.copy()
        for j in range(1, len(a)):
            a[j] = temp[j] - temp[j - 1] * x[i]
        a[0] = 1
    #print(f"hoocnerNhan: {a}") 
    return a
    
def hoocnerChia(a, value):
    b = [a[0]]
    for i in range(1, len(a)):
        b.append(a[i] + b[i - 1] * value)
    #print(f"hoocnerChia: {b}")
    return b

def noiSuyLagrange(x, y):
    a = np.zeros(len(x))
    phi = hoocnerNhan(x)
    for i in range(len(x)):
        omega = hoocnerChia(phi, x[i])
        omega.pop()
        k = y[i]/hoocnerChia(omega, x[i]).pop()
        for j in range(len(a)):
            a[j] += k * omega[j]
    print(f"NoiSuy: {a}")
    return a.tolist()
```

## 1. Nhập dữ liệu

Demo

```python
x, y = doc_input("input.xlsx")
print("Mốc nội suy x: {0}".format(x))
print("Mốc nội suy y: {0}\n".format(y))
a = noiSuyLagrange(x, y)
print("Bậc đa thức nội suy: {0}. Đa thức nội suy là:".format(len(a) - 1))
for i in range(len(a)):
    print("{0} x^{1}".format(a[i], len(a)-1-i))
```
