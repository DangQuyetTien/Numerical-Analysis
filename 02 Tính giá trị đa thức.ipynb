Tính giá trị đa thức bậc n: Q(x) tại c

```python
# Mảng a chứa các hệ số của đa thức Q(x)
# Q(x) = 2x^5 + 4x^4 + 3x^3 + 1x^2 + 6x + 8
a = [2, 4, 3, 1, 6, 8]

# Giá trị c tại đó chúng ta cần tính Q(c)
c = 5

# n là bậc của đa thức (số hệ số - 1)
n = len(a) - 1

# Khởi tạo giá trị b bằng hệ số cuối cùng của đa thức (hệ số tự do)
b = a[n]

# Vòng lặp for để áp dụng thuật toán Horner
# Lặp ngược từ hệ số gần cuối (n-1) về hệ số đầu tiên (0)
for i in range(n - 1, -1, -1):
  b = b * c + a[i]

# In kết quả cuối cùng
print(f"Giá trị của Q(x) tại c = {c} là: {b}")
```
