Chia đa thức cho x - c

```python
# P(x) = 5x^7 + 2x^6 + 4x^5 + 5x^4 + 2x^3 + 5x^2 + 1x + 8
a = [5, 2, 4, 5, 2, 5, 1, 8]  # mảng hệ số của đa thức ban đầu
c = 5                           # giá trị c từ biểu thức chia (x - c)

# Khởi tạo danh sách b để chứa các hệ số của đa thức thương và số dư
b = []

# n là bậc của đa thức (số phần tử - 1)
n = len(a) - 1

# Phần tử đầu tiên của mảng kết quả (b) luôn bằng phần tử đầu tiên của mảng ban đầu (a)
b.append(a[0])

# Vòng lặp để tính các hệ số còn lại và số dư
# lặp từ i = 1 đến n (tức là hết mảng a)
for i in range(1, n + 1):
    # Áp dụng công thức của sơ đồ Horner
    # b[i] = b[i-1]*c + a[i]
    new_b_element = b[i-1] * c + a[i]
    b.append(new_b_element)

# In kết quả
# Các hệ số của đa thức mới (thương) là các phần tử của b, trừ phần tử cuối cùng
quotient_coefficients = b[:-1]

# Số dư chính là phần tử cuối cùng của b
remainder = b[-1]

print(f"Mảng hệ số của đa thức mới là:\n{quotient_coefficients}")
print(f"Với phần dư là: {remainder}")
```
